```python
year = [0] * 250
name_eng = [' '] * 250
rating = [0] * 250
marks_number = [0] * 250
movie_id = [0] * 250
runtimes = [0]*250
cast = [0]*250
actorsBase = [0]*1000
genresBase = [0]*250

genres = [0]*250;
k = 0

for i in range(1, 5):
    movie = top[i-1]
    name_eng[i-1] = movie.get('title')
    year[i-1] = movie.get('year')
    movie_id[i-1] = movie.movieID
    rating[i-1] = movie.get('rating')
    marks_number[i-1] = movie.get('votes')
    ia.update(movie, info = ['main'])
    cast[i-1] = movie.get('cast')
    
    for j in range(len(cast[i-1])):
        cast[i-1][j] = cast[i-1][j].get('name')

    for j in range(len(cast[i-1])):
            actorsBase[k] = cast[i-1][j] #записали всех подряд актеров
            k += 1
    genres[i-1] = movie.get('genres')
```

```python
def getIndexInSet(string, currentSet):
    resultIndex = 0
    for element in currentSet:
        if (string == element):
            return resultIndex
        resultIndex += 1
```

```python
genres = [0]*100
allGenres =[0]*250
namedGenres = {}
z = 0
for i in range(1,15): 
    movie = top[i-1]
    ia.update(movie, info = ['main'])
    genres[i-1] = movie.get('genres')
    namedGenres[i-1] = genres[i-1]
    #print('fds ', genres[i-1])
    for j in range(len(genres[i-1])):
       # print('fererere r ', j, ' ', genres[i-1][j])
        allGenres[z] = genres[i-1][j]
        z += 1

genresSet = set(allGenres)
genresSet.remove(0)
# получили множество жанров
```

```python
filmVector = [[0] *( len(actorSet) + len(genresSet))for i in range(251)]
k = 0
filmRec = {}
namedFilms = {}
for i in range(1,5):
    movie = top[i-1]
    name_eng[i-1] = movie.get('title')
    ia.update(movie, info = ['main'])
    cast[i-1] = movie.get('cast')
    genres[i-1] = movie.get('genres')
    namedFilms[i-1] = name_eng[i-1]
    
    for j in range(len(cast[i-1])):
        if (cast[i-1][j] in actorSet):
            index = getIndexInSet(cast[i-1][j],actorSet)
            filmVector[i-1][index] = 1
    #filmRec[name_eng[i-1]] = filmVector[i-1]
    ###
    #k = len(filmVector) - len(cast[i-1]) +1; 
    k = 0
    for z in range(len(cast[i-1]), len(filmVector[i-1])):    
        if (k<len(genres[i-1]) and genres[i-1][k] in genresSet):
            index = getIndexInSet(genres[i-1][k], genresSet)
            filmVector[i-1][z+index] = 1
        k += 1
    filmRec[name_eng[i-1]] = filmVector[i-1]
#построили вектора для фильмов
```

```python
#ищем 20 ближайших фильмов
N = int(input())
while(N<=0 or N>250):
    N = int(input())
    
norma = {}
currentFilm = filmVector[N]
for i in range(1,5):
    if (i-1 != N-1):
        sum = 0
        for j in range(1, len(actorSet)):
            sum += (filmVector[i-1][j-1]-currentFilm[j-1])*(filmVector[i-1][j-1]-currentFilm[j-1])
            norma[get_key(filmRec,filmVector[i-1]) ] = math.sqrt(sum)
    else:
        norma[get_key(filmRec,filmVector[i-1])] = 800000
print("Norma")
print(norma)
```
